```python
# Task 1

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set()

np.random.seed(2)
x = np.linspace(0, 24, 10)
y = np.random.rand(10, 3)
plt.plot(x, y)
plt.legend(['Fri', 'Sun', 'Sat'], ncol=3, loc='lower left')

plt.title('Level of walking activity at weekends', fontsize=18)
plt.xlabel('Hours', fontsize=16)
plt.ylabel('Level of walking activity,\nnon-existing scale ', fontsize=16)
```

```python
# Task 2.

import seaborn as sns
from Bio import SeqIO
import matplotlib.pyplot as plt

print('The program draws distribution of sequence lengths derived from your fasta file')
input_fasta = input('Please, enter the path to your fasta file: ')
def draw_seq_lengths(input_fasta):
    plt.figure(figsize=(12, 8))
    plt.rcParams['font.size'] = 12
    plt.rcParams['savefig.bbox'] = 'tight'

    plt.rcParams['axes.labelpad'] = 12   # pace between label and axis
    plt.title('Distribution of sequence lengths', fontsize=18)
    plt.xlabel('Length of sequence', fontsize=16)
    plt.ylabel('Number of sequences with certain length', fontsize=16)
    sns.histplot(list(map(len, SeqIO.parse(input_fasta, 'fasta'))), color = "dodgerblue", 
                 bins = range(min(map(len, SeqIO.parse(input_fasta, 'fasta'))), max(map(len, SeqIO.parse(input_fasta, 'fasta')))), 
                kde = True)
draw_seq_lengths(input_fasta)
```

```python
# Task 3

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="dark")

# Data simulation from a bivariate Gaussian
n = 2000
mean = [0, 0]
cov = [(7, 0.5), (0.4, .2)]
rng = np.random.RandomState(0)
x, y = rng.multivariate_normal(mean, cov, n).T

# Draw histogram and scatterplot with density contours
fig, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=8, color=".17")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap='cubehelix')
sns.kdeplot(x=x, y=y, levels=5, color="y", linewidths=1)
```
